```python
# import pandas as pd
import numpy as np
import pandas as pd
import datetime as dt
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as dates
from mpl_finance import candlestick_ohlc
import mpl_toolkits
import json
import requests

# This line is necessary for the plot to appear in a Jupyter notebook
%matplotlib inline
# Control the default size of figures in this Jupyter notebook
%pylab inline

def stock_url(function, symbol, outputsize):
    API_KEY = "WWLN1J8UJ2I7Q8ML"
    url = (f"https://www.alphavantage.co/query?function={function}&symbol={symbol}&outputsize={outputsize}&apikey={API_KEY}")
    return url

ema_url1 = "https://www.alphavantage.co/query?function=EMA&symbol=MSFT&interval=daily&time_period=20&series_type=open&apikey=WWLN1J8UJ2I7Q8ML"
ema_url2 = "https://www.alphavantage.co/query?function=EMA&symbol=MSFT&interval=daily&time_period=50&series_type=open&apikey=WWLN1J8UJ2I7Q8ML"


#api url and key stored as variables
#_API_URL = "http://www.alphavantage.co/query?"
#_SYMBOL = "BA"
#_FUNCTION_VALUE = "TIME_SERIES_DAILY_ADJUSTED"
```

```python
#request api of current daily stock of selected symbol
req_url = stock_url("TIME_SERIES_DAILY_ADJUSTED", "MSFT", "full")
print(req_url)

print(ema_url1)
print(ema_url2)
```

```python
#convert response to python json list
data = requests.get(req_url).json()['Time Series (Daily)']
dataTech1 = requests.get(ema_url1).json()['Technical Analysis: EMA']
dataTech2 = requests.get(ema_url2).json()['Technical Analysis: EMA']
```

```python
#convert list to dataframe
df = pd.DataFrame(columns=['Date','Open','High','Low','Close'])
for k,v in data.items():
    date = dt.datetime.strptime(k, '%Y-%m-%d')
    data_row = [date.date(),float(v['1. open']),float(v['2. high']),
                float(v['3. low']),float(v['4. close'])]
    df.loc[-1,:] = data_row
    df.index = df.index + 1
```

```python
df.head()
```

```python
dfTech1 = pd.DataFrame(columns=['Date','EMA'])
for k,v in dataTech1.items():
    date = dt.datetime.strptime(k, '%Y-%m-%d')
    data_row = [date.date(),float(v['EMA'])]
    dfTech1.loc[-1,:] = data_row
    dfTech1.index = dfTech1.index + 1
```

```python
dfTech2 = pd.DataFrame(columns=['Date','EMA'])
for k,v in dataTech2.items():
    date = dt.datetime.strptime(k, '%Y-%m-%d')
    data_row = [date.date(),float(v['EMA'])]
    dfTech2.loc[-1,:] = data_row
    dfTech2.index = dfTech2.index + 1
```

```python
df = df.sort_values('Date', ascending=False)
df.tail()
```

```python
dfTech1 = dfTech1.sort_values('Date', ascending=False)
dfTech1.head()
```

```python
dfTech2 = dfTech2.sort_values('Date', ascending=False)
dfTech2.tail()
```

```python
df = df.iloc[:500]
dfTech1 = dfTech1.iloc[:500]
dfTech2 = dfTech2.iloc[:500]
```

```python
fig, ax = plt.subplots(figsize=(18, 9))
ax.set_title('Adjusted close value', fontdict={'fontsize': 16, 'fontweight': 'medium'})
ax.plot_date(df['Date'], df['Open'], 'r-', label='OPEN')
ax.plot_date(df['Date'], df['Close'], 'b-', label='CLOSE')
#ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=(MO)))
#ax.xaxis.set_minor_formatter(dates.DateFormatter('%d\n%a'))
ax.xaxis.grid(True, which="minor")
ax.yaxis.grid()
ax.xaxis.set_major_locator(dates.YearLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n\n%Y'))
plt.legend(loc=2)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.set_title('Adjusted close value', fontdict={'fontsize': 16, 'fontweight': 'medium'})
ax.plot_date(dfTech1['Date'], dfTech1['EMA'], 'r-', label='EMA fast')
ax.plot_date(dfTech2['Date'], dfTech2['EMA'], 'b-', label='EMA slow')
#ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=(MO)))
#ax.xaxis.set_minor_formatter(dates.DateFormatter('%d\n%a'))
ax.xaxis.grid(True, which="minor")
ax.yaxis.grid()
ax.xaxis.set_major_locator(dates.YearLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n\n%Y'))
plt.legend(loc=2)
plt.tight_layout()
plt.show()
```

```python
amp_prev = 0
profit = 0
entry = 0
exit = 0
counter = 0
buy = True #restriction
for i in df['Date']:
    #get EMA10 value
    ema10 = float(dfTech1.loc[dfTech1.Date == i].EMA)
    #get EMA20 value
    ema20 = float(dfTech2.loc[dfTech2.Date == i].EMA)
    #get amplitude/difference between EMA10 & EMA20
    amp = ema10 - ema20
    current = float(df.loc[df.Date == i].Open)
    if amp > 0:
        if entry == 0:
            entry = current
            amp_prev = amp
            buy = True
        elif buy == True: #restrict reentry
            continue
        else:
            exit = current
            profit += (entry - exit)
            counter += 1
            entry = current
    elif amp < 0:
        if entry == 0:
            entry = current
            amp_prev = amp
            buy = False
        elif buy == False: #restrict reentry
            continue
        else:
            exit = current
            profit += (exit - entry)
            counter += 1
            entry = current
print(profit)
print(counter)
```
